````python
import pandas as pd
import glob
import os
````

````python
# Verificar que existen los archivos CSV descargados
archivos = glob.glob('data/ecommerce_*.csv')
if not archivos:
    print("❌ No se encontraron los archivos. Asegurate de descargarlos en la carpeta data/")
    print("   Deberías tener: ecommerce_orders.csv, ecommerce_customers.csv, etc.")
else:
    print(f"📂 Archivos encontrados: {len(archivos)}")
    for f in sorted(archivos):
        print(f"  - {os.path.basename(f)}")
````

````python
# Cargar los CSVs principales
df_orders = pd.read_csv('data/ecommerce_orders.csv')
df_order_items = pd.read_csv('data/ecommerce_order_items.csv')
df_customers = pd.read_csv('data/ecommerce_customers.csv')
df_products = pd.read_csv('data/ecommerce_products.csv')
````

````python
# Explorar
print("📈 Resumen:")
print(f"\nEl dataset de órdenes tiene {len(df_orders)} filas y {len(df_orders.columns)} columnas")
print(f"\nEl dataset de items tiene {len(df_order_items)} filas y {len(df_order_items.columns)} columnas")
print(f"\nEl dataset de clientes tiene {len(df_customers)} filas y {len(df_customers.columns)} columnas")
print(f"\nEl dataset de productos tiene {len(df_products)} filas y {len(df_products.columns)} columnas")

print("\nPrimeras filas de órdenes\n")
print(df_orders.head(20))
print("\nEstadísticas principales de órdenes\n")
print(df_orders.describe())
````

````python
# Chequando nulos
print("Nulos de órdenes\n")
print(df_orders.isnull().sum())

print("\nNulos de ítems\n")
print(df_order_items.isnull().sum())

print("\nNulos de productos\n")
print(df_products.isnull().sum())

print("\nNulos de clientes\n")
print(df_customers.isnull().sum())

df_orders['promotion_id'] = df_orders['promotion_id'].fillna(0)
df_orders['notes'] = df_orders['notes'].fillna("No notes.")

print("\nNulos de órdenes tras el arreglo\n")
print(df_orders.isnull().sum())
````

````python
# Chequeamos si hay duplicados

print("Analizando duplicados\n")
print(f"\nÓrdenes: {df_orders.duplicated().sum()}")
print(f"\nÍtems: {df_order_items.duplicated().sum()}")
print(f"\nClientes: {df_customers.duplicated().sum()}")
print(f"\nProductos: {df_products.duplicated().sum()}")

print("\nAnalizando duplicados importantes")
ordenes_duplicadas = df_orders.duplicated(subset=['order_id']).sum()
print(f"\nÓrdenes duplicadas {ordenes_duplicadas}")

print(f"\nTamaño del dataset órdenes con duplicados: {len(df_orders)}")
df_orders_clean = df_orders.drop_duplicates()
print(f"\nTamaño del dataset órdenes sin duplicados: {len(df_orders_clean)}")
````

````python
# Chequeamos si las columnas de órdenes son del tipo correcto
print(df_orders_clean.dtypes)

df_orders_clean['order_date'] = pd.to_datetime(df_orders_clean['order_date']) # order_date debe ser una fecha

print(f"\n Ahora \n{df_orders_clean.dtypes}")
````

````python
# Obtenemos a los 5 mejores clientes
best_clients = df_orders_clean.groupby('customer_id').agg({'total_amount': 'sum', 'customer_id': 'count'})
best_clients = best_clients.rename(columns={'total_amount': 'total gastado', 'customer_id': 'cantidad de compras'})
best_clients = best_clients.sort_values('total gastado',ascending=False).reset_index()
best_clients = best_clients.rename(columns={'customer_id': 'Cliente ID'})
print("🏆 Top 5 clientes:\n")
print(best_clients.head(5))
````

````python
# Obtenemos el producto más vendido
best_products = df_order_items.groupby('product_id')['quantity'].sum().sort_values(ascending=False)
best_product_id = best_products.idxmax()
best_product_quantity = best_products.max()
best_product_name = df_products[df_products['product_id'] == best_product_id]['product_name'].iloc[0]

print(f"El producto más vendido es {best_product_name} con id {best_product_id} y un total de {best_product_quantity} ventas.")
````

````python
df_orders_clean['mes_año'] = df_orders_clean['order_date'].dt.to_period('M')
ventas_x_mes = df_orders_clean.groupby('mes_año')['total_amount'].sum().reset_index().rename(columns={'mes_año': 'fecha', 'total_amount': 'ventas'})
print(ventas_x_mes)
````

````python
# Crear carpeta output si no existe
import os
os.makedirs('output', exist_ok=True)

# Guardar métricas en CSV
best_clients.to_csv('output/ventas_por_cliente.csv', index=True)
ventas_x_mes.to_csv('output/ventas_por_mes.csv', index=False)

# Guardar datos limpios
df_orders_clean.to_csv('output/orders_clean.csv', index=False)

print("✅ Archivos CSV guardados en output/")
````

````python
# Crear carpeta output si no existe
import os
os.makedirs('output', exist_ok=True)

# Guardar métricas en CSV
best_clients.to_parquet('output/ventas_por_cliente.parquet', index=True)
ventas_x_mes.to_parquet('output/ventas_por_mes.parquet', index=False)

# Guardar datos limpios
df_orders_clean.to_parquet('output/orders_clean.parquet', index=False)

print("✅ Archivos CSV guardados en output/")
````

````python
# Comparar tamaños
csv_size = os.path.getsize('output/orders_clean.csv') / 1024
parquet_size = os.path.getsize('output/orders_clean.parquet') / 1024

print(f"Tamaño CSV: {csv_size:.1f} KB")
print(f"Tamaño Parquet: {parquet_size:.1f} KB")
print(f"Parquet es {csv_size/parquet_size:.1f}x más chico")
````
